```python
# Import modules
import pandas as pd
import requests
import datetime as dt
import time
# Ignore the SettingWithCopyWarning
pd.options.mode.chained_assignment = None
```

```python
# Read commercial real estae data into DataFrame
comm_re_df = pd.read_csv('comm_re_data.csv')
comm_re_df
```

```python
# Rename columns of and convert 'Date' column to datetime format
comm_re_df['Date'] = pd.to_datetime(comm_re_df['Date'])
comm_re_df.sort_values(by=['Date'],inplace=True)
comm_re_df.head(10)
```

```python
# Lookup dates from 11-01-2019 to 03-01-2020
comm_re_df_lkup = (comm_re_df['Date'] > '2017-12-30') & (comm_re_df['Date'] <= '2020-04-01')
comm_re_df = comm_re_df.loc[comm_re_df_lkup]
comm_re_df.head(10)
```

```python
# Reset the index and drop the index column
comm_re_df.reset_index(inplace=True)
comm_re_df.drop(columns=['index'], inplace=True)
comm_re_df.set_index(['Date'], inplace=True)
comm_re_df.head(10)
```

```python
# Export cleaned data to csv
comm_re_df.to_csv('comm_re_clean.csv')
```
