```python
# Download csv from URL and allow 10 seconds before moving to the next cell
# for file to update
url = "https://raw.githubusercontent.com/nytimes/covid-19-data/master/us.csv"
file_request = requests.get(url, allow_redirects=True)
with open("covid_cases_us.csv", "wb") as file:
    file.write(file_request.content)
time.sleep(10) 
```

```python
# Read csv into DataFrame
covid_df = pd.read_csv(
    'covid_cases_us.csv',
    index_col='date',
    parse_dates=True,
    infer_datetime_format=True
)
covid_df.head()
```

```python
# sort index in ascending order
covid_df.sort_index(inplace=True)
```

```python
# Lookup dates between January 21, 2020 and April 30, 2020
# then set the index to 'date'
covid_df = covid_df.loc['2020-01-21':'2020-04-30',:]
covid_df.head()
```

```python
covid_df.tail()
```

```python
# drop unnecessary columns
covid_df.drop(columns='deaths', inplace=True)
covid_df.head()
```

```python
# concatenate 19-20 cumulative returns and covid data for analysis
covid_sp500_df = pd.concat(
    [sp500_df, covid_df],
    axis='columns',
    join='inner',
)
covid_sp500_df.head()
```

```python
covid_sp500_df.tail()
```

```python
# https://plotly.com/python/multiple-axes/
# plotly multiple axes plot:

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x=covid_sp500_df.index, y=covid_sp500_df['S&P 500'], name="S&P 500"),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=covid_sp500_df.index, y=covid_sp500_df['cases'], name="COVID-19 cases"),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title_text="S&P 500 closing prices vs. COVID-19 cases"
)

# Set x-axis title
fig.update_xaxes(title_text="Date")

# Set y-axes titles
fig.update_yaxes(title_text="S&P 500", secondary_y=False)
fig.update_yaxes(title_text="COVID-19 cases", secondary_y=True)

fig.show()
```

```python
rapid_api_key = os.getenv('X_RAPIDAPI_KEY')
```

```python
type(rapid_api_key)
```

```python
# set api url
url = 'https://apidojo-yahoo-finance-v1.p.rapidapi.com/stock/v2/get-historical-data'

# set query for api - ^NDXT = nasdaq tech 100 index, representative of tech sector
querystring = {'frequency':'1d','filter':'history','period1':'1272585600','period2':'1588204800','symbol':'%5ENDXT'}

# set headers for api request
headers = {
    'x-rapidapi-host': 'apidojo-yahoo-finance-v1.p.rapidapi.com',
    'x-rapidapi-key': rapid_api_key,
    }

# get response from api request
response = requests.request('GET', url, headers=headers, params=querystring)
```

```python
tech_df = pd.io.json.json_normalize(response.json()['prices'])
```

```python
tech_df.head()
```

```python
# convert date column to datetime from epoch time
tech_df['date'] = pd.to_datetime(tech_df['date'], unit='s')
# set date as index column
tech_df.set_index('date', inplace=True)
# remove timestamp from index
tech_df.index = tech_df.index.normalize()

tech_df.head()
```

```python
# note to self:
# to remove timestamps from datetime columns that are not in the index,
# use _df['date'].dt.normalize() instead of _df['date'].normalize()
```

```python
# remove unnecessary columns
tech_df.drop(columns=['adjclose', 'high', 'low', 'open', 'volume'], inplace=True)

# rename close column to tech_close
# since there's only one column, we can use the .columns property rather than
# the .replace() method
tech_df.columns = ['tech_close']

# sort index in ascending order
tech_df.sort_index(inplace=True)

tech_df.head()
```

```python
# calculate tech daily returns
tech_returns_df = tech_df.pct_change()
tech_returns_df.head()
```

```python
# drop NaN's
tech_returns_df.dropna(inplace=True)
tech_returns_df.head()
```

```python
tech_returns_df.plot(figsize=(20,10))
```

```python
# calculate tech cumulative returns
tech_cumulative_returns = (tech_returns_df + 1).cumprod()

tech_cumulative_returns.head()
```

```python
tech_cumulative_returns.plot(figsize=(20,10))
```

```python
# concatenate tech cumulative returns, s&p500 cumulative returns
# and covid data for analysis
combined_df = pd.concat(
    [tech_df, covid_sp500_df],
    axis='columns',
    join='inner',
)
combined_df.head()
```

```python
# rename S&P 500 column to sp500_close and cases to covid_cases
combined_df.rename(columns={
    'S&P 500':'sp500_close',
    'cases':'covid_cases'
}, inplace=True)
combined_df.head()
```

```python
# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x=combined_df.index, y=combined_df['sp500_close'],
               name='S&P 500 price'),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=combined_df.index, y=combined_df['tech_close'],
               name='Tech price'),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=combined_df.index, y=combined_df['covid_cases'],
               name="COVID-19 cases"),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title_text="S&P 500 & Tech sector closing prices vs. COVID-19 cases"
)

# Set x-axis title
fig.update_xaxes(title_text="Date")

# Set y-axes titles
fig.update_yaxes(title_text="Market prices", secondary_y=False)
fig.update_yaxes(title_text="COVID-19 cases", secondary_y=True)

fig.show()
```

```python
# calculate returns for s&p500 in combined_df
combined_df['sp500_daily_returns'] = combined_df['sp500_close'].pct_change()

# calculate returns for tech in combined_df
combined_df['tech_daily_returns'] = combined_df['tech_close'].pct_change()

combined_df.head()
```

```python
# calculate cumulative returns for s&p500 in combined_df
combined_df['sp500_cumulative_returns'] = (
    (combined_df['sp500_daily_returns'] + 1).cumprod()
)

# calculate cumulative returns for tech in combined_df
combined_df['tech_cumulative_returns'] = (
    (combined_df['tech_daily_returns'] + 1).cumprod()
)

combined_df.head()
```

```python
# rearrange columns
combined_df = combined_df[
    [
    'covid_cases',
    'sp500_close',
    'sp500_daily_returns',
    'sp500_cumulative_returns',
    'tech_close',
    'tech_daily_returns',
    'tech_cumulative_returns'
    ]
]
combined_df.head()
```

```python
# drop NaN's
combined_df.dropna(inplace=True)
combined_df.head()
```

```python
# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x=combined_df.index, y=combined_df['sp500_cumulative_returns'],
               name='S&P 500 cumulative returns'),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=combined_df.index, y=combined_df['tech_cumulative_returns'],
               name='Tech cumulative returns'),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=combined_df.index, y=combined_df['covid_cases'],
               name="COVID-19 cases"),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title_text="S&P 500 & Tech sector cumulative returns vs. COVID-19 cases"
)

# Set x-axis title
fig.update_xaxes(title_text="Date")

# Set y-axes titles
fig.update_yaxes(title_text="Market cumulative returns", secondary_y=False)
fig.update_yaxes(title_text="COVID-19 cases", secondary_y=True)

fig.show()
```

```python
# calculate 30-day rolling correlation between S&P500 prices and COVID-19 cases,
# between S&P500 prices and tech sector prices,
# and between tech sector prices and COVID-19 cases

corr_df = combined_df['covid_cases'].rolling(window=30).corr(other=combined_df[
    ['sp500_close',
     'tech_close'
    ]])

# drop NaN's
corr_df.dropna(inplace=True)

corr_df.head(9)
```

```python
sb.heatmap(corr_df)
```

```python
corr_df.hvplot()
```

```python
# create empty dataframe
risk_df = pd.DataFrame()

# add covid cases to dataframe
risk_df['covid_cases'] = combined_df['covid_cases']

# calculate s&p500 standard deviation
risk_df['sp500_std_dev'] = combined_df['sp500_daily_returns'].rolling(window=30).std(ddof=0)

# calculate tech sector standard deviation
risk_df['tech_std_dev'] = combined_df['tech_daily_returns'].rolling(window=30).std(ddof=0)

# drop NaN's
risk_df.dropna(inplace=True)

risk_df.head()
```

```python
# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x=combined_df.index, y=risk_df['sp500_std_dev'],
               name='S&P 500 standard deviation'),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=combined_df.index, y=risk_df['tech_std_dev'],
               name='Tech standard deviation'),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=combined_df.index, y=risk_df['covid_cases'],
               name="COVID-19 cases"),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title_text="S&P 500 & Tech sector risk vs. COVID-19 cases"
)

# Set x-axis title
fig.update_xaxes(title_text="Date")

# Set y-axes titles
fig.update_yaxes(title_text="Market risk", secondary_y=False)
fig.update_yaxes(title_text="COVID-19 cases", secondary_y=True)

fig.show()
```
