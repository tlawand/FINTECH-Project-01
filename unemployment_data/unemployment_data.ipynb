```python
# Import modules
import pandas as pd
import datetime as dt
import hvplot.pandas
```

```python
# Set DataFrame
unrate_df = pd.read_csv('./UNRATENSA.csv')
unrate_df.head(10)
```

```python
# Rename columns and convert 'Date' column to datetime format
unrate_df.rename(columns={'DATE':'Date', 'UNRATENSA':'unrate'}, inplace=True)
unrate_df['Date'] = pd.to_datetime(unrate_df['Date'])
unrate_df.sort_values(by=['Date'],inplace=True)
unrate_df.head(10)
```

```python
# Lookup dates from 04-30-2010 to 03-01-2020
unrate_df_lkup = (unrate_df['Date'] > '2010-04-30') & (unrate_df['Date'] <= '2020-03-01')
unrate_df = unrate_df.loc[unrate_df_lkup]
unrate_df.head(10)
```

```python
# Reset the index and drop the index column
unrate_df.reset_index(inplace=True)
unrate_df.drop(columns=['index'], inplace=True)
unrate_df.set_index(['Date'], inplace=True)
unrate_df.head(10)
```

```python
# Plot the DataFrame
unrate_df.hvplot(height=500, width=1200)
```
